```python
# See the first 3 raw rows
df[0].head(3)

```

```python
# Split on whitespace
df_split = df[0].str.split(expand=True)

# Check the number of columns
print("Number of columns:", df_split.shape[1])

# Preview the data
df_split.head()
```

```python
df_split.columns = [
    'CRIM',        # Crime rate
    'ZN',          # Residential land zoned
    'INDUS',       # Non-retail business acres per town
    'CHAS',        # Charles River dummy variable (1 if bounds river)
    'NOX',         # Nitric oxide concentration
    'RM',          # Average number of rooms per dwelling
    'AGE',         # Proportion of owner-occupied units built before 1940
    'DIS',         # Distances to employment centres
    'RAD',         # Accessibility to highways
    'TAX',         # Property-tax rate
    'PTRATIO',     # Pupil-teacher ratio
    'B',           # Proportion of Black population
    'LSTAT',       # % lower status of the population
    'MEDV'         # Median house value (target)
]
```

```python
df_split = df_split.apply(pd.to_numeric)
df_split.info()
```

```python
plt.hist(df_split['MEDV'], bins=30, color='skyblue')
plt.title("Distribution of House Prices")
plt.xlabel("Price (in $1000s)")
plt.ylabel("Number of Houses")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df_split['RM'], df_split['MEDV'], alpha=0.6, color='purple')
plt.title("Price vs Average Number of Rooms")
plt.xlabel("Average Rooms per Dwelling (RM)")
plt.ylabel("Median House Price (MEDV)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_split.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Boston Housing Features")
plt.show()
```

## 📝 Summary Insights (EDA on Boston Housing Dataset)

1. **Price Distribution**: Most houses are priced between $20,000 and $25,000, with a few high-value outliers.
2. **Room Count & Price**: There's a positive correlation between number of rooms (`RM`) and price (`MEDV`), meaning more rooms generally indicate higher price.
3. **Negative Impact Factors**:
   - `LSTAT` (% lower income population) shows strong **negative correlation** with price (r = -0.74).
   - `CRIM` (crime rate) also has a moderate negative effect.
4. **Strongest Positive Correlations**:
   - `RM` (rooms) shows a strong **positive correlation** with price (r = +0.7).
   - `B` (Black population proportion) also correlates slightly positively.
5. **No Missing Values**: Data is clean and complete.

✅ This EDA gives a clear overview of how housing features affect prices in Boston.
